```python
#outliers treatment for movies
# Extract numeric part of Duration
df["Duration_num"] = df["Duration"].str.extract(r"(\d+)").astype(float)
df["Duration_unit"] = df["Duration"].str.extract(r"([a-zA-Z]+)")

# Outlier treatment only for Movies with Minutes
movies = df[(df["Type"] == "Movie") & (df["Duration_unit"].str.contains("min", case=False, na=False))]

q1 = movies["Duration_num"].quantile(0.25)
q3 = movies["Duration_num"].quantile(0.75)
iqr = q3 - q1
lower = q1 - 1.5 * iqr
upper = q3 + 1.5 * iqr

df.loc[movies.index, "Duration_num_capped"] = movies["Duration_num"].clip(lower, upper)
print("\n--- Outlier treatment applied for Movie durations ---")
```

```python
#duplicate removal
before = df.shape[0]
df = df.drop_duplicates()
after = df.shape[0]
print(f"\nDuplicates removed: {before - after}")
```

```python
#feature engineering
# Convert Release_Date to datetime and extract Year

df.loc[:, "Release_Date"] = pd.to_datetime(df["Release_Date"], errors="coerce")
df.loc[:, "Year"] = df["Release_Date"].dt.year.fillna(0).astype(int)
```

```python

# Extract Primary Genre (first category listed)
df = df.assign(Primary_Genre = df["Category"].astype(str).apply(lambda x: x.split(",")[0]))

```

```python
# Extract Year
df.loc[:, "Year"] = df["Release_Date"].dt.year.fillna(0).astype(int)
```

```python
# Extract Primary Genre
df.loc[:,"Primary_Genre"] = df["Category"].apply(lambda x: x.split(",")[0])
```

```python
# Extract Primary Country safely
df.loc[:, "Primary_Country"] = df["Country"].apply(
    lambda x: str(x).split(",")[0] if str(x) != "Not Available" else "Not Available"
)

# Show results
print("\n--- Feature Engineering Completed ---")
print(df[["Title", "Type", "Year", "Primary_Genre", "Primary_Country", "Duration_num_capped"]].head())
```
